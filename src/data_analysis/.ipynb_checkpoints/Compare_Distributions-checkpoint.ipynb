```python
fig, ax = plt.subplots(8, 2, figsize=(30, 80))

count = 0
for race in races:
    raceID = race['raceID']
    
    x = (int)(count / 2)
    y = count % 2
    ax[x, y].plot(drivers['Names'], distr_official[raceID]*100, '--o', label='official voting')
    ax[x, y].plot(drivers['Names'], distr_twitter[raceID]*100, '--o', label='twitter voting')
    ax[x, y].set_title(raceID)
    ax[x, y].set_ylabel('percentual voting')
    ax[x, y].legend()
    plt.setp( ax[x, y].xaxis.get_majorticklabels(), rotation=70 )
    
    count += 1

fig.tight_layout(pad=2.0)
plt.show()
```
